A simple notebook to inspect the available data. Here’s what we are supposed to have:

**Files:**
- `customers.csv` - customer descriptions
- `receipts.parquet` - purchase history
- `campaigns.csv` - campaign history

**Field descriptions:**
- `customers.csv`
    - `customer_id` - customer ID
    - `age` - customer age
    - `location` - place of residence

- `receipts.parquet`
    - `customer_id` - customer ID
    - `date` - purchase date
    - `purchase_amt` - amount of fishtekh purchased (in grams)
    - `purchase_sum` - purchase cost (in orens)

Note that a customer **does not buy fishtekh more than once per day**.

- `campaigns.csv`
    - `customer_id` - customer ID
    - `date` - first day the discount is active
    - `n_offer_days` - discount duration (in days)
    - `target_group_flag` - target (1) / control (0) group flag

Overall - that's about right, with the exception of "Unnamed: 0" columns. Let's save the files without them:

```python
# Re-load the data
campaigns = pd.read_csv("data/campaigns.csv")
customers = pd.read_csv("data/customers.csv")
receipts = pd.read_parquet("data/receipts.parquet")

# Remove accidental index columns if present, then overwrite source files
campaigns = campaigns.drop(columns=["Unnamed: 0"], errors="ignore")
customers = customers.drop(columns=["Unnamed: 0"], errors="ignore")
receipts = receipts.drop(columns=["Unnamed: 0"], errors="ignore")

campaigns.to_csv("data/campaigns.csv", index=False)
customers.to_csv("data/customers.csv", index=False)
receipts.to_parquet("data/receipts.parquet", index=False)
```
